## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('Task (2) Dataset.csv')
```

### Identify and analyze patterns in the data.

```python
print("Current Data Types:")
print(df.dtypes)

df['Datetime'] = pd.to_datetime(df['Datetime'])
df['City'] = df['City'].astype('category')
df['TrafficIndexLive'] = df['TrafficIndexLive'].astype(int)
df['JamsCount'] = df['JamsCount'].astype(int)
df['JamsDelay'] = df['JamsDelay'].astype(float)
df['JamsLength'] = df['JamsLength'].astype(float)
df['TrafficIndexWeekAgo'] = df['TrafficIndexWeekAgo'].astype(int)
df['TravelTimeHistoric'] = df['TravelTimeHistoric'].astype(float)
df['TravelTimeLive'] = df['TravelTimeLive'].astype(float)

print("\nUpdated Data Types:")
print(df.dtypes)
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TrafficIndexLive'], bins=20, kde=True, color='blue')
plt.title('Distribution of Traffic Index Time')
plt.xlabel('Traffic Index Live')
plt.ylabel('Date')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
# TrafficIndexLive vs. JamsCount
plt.figure(figsize=(10, 6))
sns.scatterplot(x='JamsCount', y='TrafficIndexLive', data=df, color='blue')
plt.title('Traffic Index Live vs. Jams Count')
plt.xlabel('Jams Count')
plt.ylabel('Traffic Index Live')
plt.show()

# TrafficIndexLive vs. JamsDelay
plt.figure(figsize=(10, 6))
sns.scatterplot(x='JamsDelay', y='TrafficIndexLive', data=df, color='green')
plt.title('Traffic Index Live vs. Jams Delay')
plt.xlabel('Jams Delay (minutes)')
plt.ylabel('Traffic Index Live')
plt.show()

# TrafficIndexLive vs. JamsLength
plt.figure(figsize=(10, 6))
sns.scatterplot(x='JamsLength', y='TrafficIndexLive', data=df, color='red')
plt.title('Traffic Index Live vs. Jams Length')
plt.xlabel('Jams Length (km)')
plt.ylabel('Traffic Index Live')
plt.show()

# JamsCount vs. JamsDelay
plt.figure(figsize=(10, 6))
sns.scatterplot(x='JamsCount', y='JamsDelay', data=df, color='purple')
plt.title('Jams Count vs. Jams Delay')
plt.xlabel('Jams Count')
plt.ylabel('Jams Delay (minutes)')
plt.show()

# JamsCount vs. JamsLength
plt.figure(figsize=(10, 6))
sns.scatterplot(x='JamsCount', y='JamsLength', data=df, color='orange')
plt.title('Jams Count vs. Jams Length')
plt.xlabel('Jams Count')
plt.ylabel('Jams Length (km)')
plt.show()

# JamsDelay vs. JamsLength
plt.figure(figsize=(10, 6))
sns.scatterplot(x='JamsDelay', y='JamsLength', data=df, color='brown')
plt.title('Jams Delay vs. Jams Length')
plt.xlabel('Jams Delay (minutes)')
plt.ylabel('Jams Length (km)')
plt.show()


corr_matrix = df[['TrafficIndexLive', 'JamsCount', 'JamsDelay', 'JamsLength']].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
